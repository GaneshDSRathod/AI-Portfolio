## Logical Operators

```python
### A             B               A and B True/ False
#   0             0               False
#   0             1               False
#   1             0               False
#   1             1               True

# Only Both true are true, if one false then always false
```

```python
### A             B               A or B True/ False
#   0             0               True
#   0             1               True
#   1             0               True
#   1             1               False

# Any one must be valid
```

```python
#Not operator Reverse the result, returns False if the result is true
```

## Bitwise Operators
### Bitwise operators are used to compare (binary) numbers

```python
m = 10  #1010     010100  101000          1,07 timestamp
n = 4  # 0100
#128 64 32 16 8 4 2 1 
```

```python
print(m & n)     #Sets each bit to 1 if both bits are 1  ---And operator
```

```python
print(m | n)     #Sets each bit to 1 if one of two bits is 1     --- OR operator
```

```python
print(m ^ n)    #Sets each bit to 1 if only one of two bits is 1  ---XOR operator ---only if both bits are different, if both bits match then 0
```

```python
print(~m)     #Inverts all the bits 
```

```python
customer = input("Customer Name: ")    #Input name

price = float(input("Product Price: "))   #price converted to float
quantity = int(input("Quantity: "))      # quantity to int

subtotal = price * quantity               #subtotal which is multiplied by price and quantity

discount = 0                               #discount defined

if subtotal >= 5000:                      # if statement when sub total is above 5000 then discount will be applied
    discount = subtotal * 0.10

final_amount = subtotal - discount         # final amount which can be discounted 

print("\n----- BILL -----")                # print

print(f"Customer : {customer}")            #customer name
print(f"Subtotal : ₹{subtotal}")           #subtotal displayed
print(f"Discount : ₹{discount}")           # discount if added
print(f"Final Amount : ₹{final_amount}")  #final amount displayed

print(f"\nEligible For Discount : {subtotal >= 5000}")   #if eligible for disount
```
