## Small Project - Course enrollment

```python
name = input("Enter your full name: ")           ## Will take input of the student 
age = int(input("Enter your age: "))             ## Take input of age of the student, while typecast it into integar
course = input("Enter your course name: ")       ## Input of Name of the course
fees = float(input("Enter course fees: "))       ## Course fees input, while typecasting it into float

first_letter = name[0]                           ## Display first letter of var name
last_letter = name[-1]                           ## Display last letter of var name

discount = fees * 0.10                           ## Discount will be applied on fees var
final_fees = fees - discount                     ## Final fees

print("\n----- STUDENT REPORT -----")            ## print statement with \n i.e new line

print(f"Name : {name.title()}")                  ## f string to print adding var, while name will be titled (uppercase first letter)
print(f"Age : {age}")                            ## age will be displayed
print(f"Course : {course.upper()}")              ## f string while course name will be titled (uppercase first letter)

print(f"\nFirst Letter : {first_letter}")        ## print with new line and f string
print(f"Last Letter : {last_letter}")

print(f"\nOriginal Fees : ₹{fees}")
print(f"Discount : ₹{discount}")
print(f"Final Fees : ₹{final_fees}")

print("\n----- DATA TYPES -----")   
print(type(name))                               ## know the type of var
print(type(age))
print(type(fees))

print(f"\nWelcome {name.title()}")
```
